```python
# Import necessary libraries
from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import warnings

# Suppress ConvergenceWarning from MLPClassifier for cleaner output
warnings.filterwarnings('ignore')

# Load the dataset
df = pd.read_csv('student-por.csv', sep=';')

# Data Preprocessing
X = df.drop('G3', axis=1)
y = df['G3']
categorical_cols = X.select_dtypes(include=['object']).columns
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# Define both classification targets
y_pass = (y >= 10).astype(int)  # Pass/Fail (1 if G3 >= 10)
y_dropout = (y <= 5).astype(int)  # Dropout (1 if G3 <= 5)

# --- Nested Cross-Validation for Both Classification Tasks ---

# Define a dictionary to hold our models for easy iteration
models = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Naive Bayes': GaussianNB(),
    'ANN': MLPClassifier(random_state=42)
}

# Define the hyperparameter search space for each model
param_grids = {
    'Logistic Regression': {'clf__C': [0.1, 1, 10, 100], 'clf__max_iter': [300, 500]},
    'Decision Tree': {'clf__max_depth': [None, 5, 10, 20]},
    'Random Forest': {'clf__n_estimators': [100, 200, 300], 'clf__max_depth': [None, 10, 20]},
    'Naive Bayes': {},
    'ANN': {'clf__hidden_layer_sizes': [(50,), (100,)], 'clf__alpha': [0.0001, 0.001], 'clf__max_iter': [300, 500]}
}

def run_classification_task(X, y, task_name):
    results_c = {}
    print(f"\nStarting Nested Cross-Validation for {task_name} Prediction...")
    print("\n" + "="*50)
    print(f"  {task_name.upper()} MODEL PERFORMANCE  ")
    print("="*50)

    # Outer K-Fold Cross-Validation for evaluation
    outer_skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

    for name, model in models.items():
        print(f"\n--- Training and Tuning {name} for {task_name} ---")

        # Inner K-Fold Cross-Validation for hyperparameter tuning
        inner_skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

        # Create an imbalance-aware pipeline
        pipeline = ImbPipeline([
            ('scaler', StandardScaler()),
            ('pca', PCA(n_components=0.95)),
            ('smote', SMOTE(random_state=42)),
            ('clf', model)
        ])

        # Lists to store metrics for each outer fold
        accuracies, precisions, recalls, f1_scores = [], [], [], []
        all_preds, all_tests = np.array([]), np.array([])

        # Outer Cross-Validation Loop
        for fold, (train_index, test_index) in enumerate(outer_skf.split(X, y)):
            X_train, X_test = X.iloc[train_index], X.iloc[test_index]
            y_train, y_test = y.iloc[train_index], y.iloc[test_index]

            # Inner Loop: Find best hyperparameters using RandomizedSearchCV
            if param_grids[name]:
                random_search = RandomizedSearchCV(
                    estimator=pipeline,
                    param_distributions=param_grids[name],
                    n_iter=5,
                    cv=inner_skf,
                    scoring='f1',
                    n_jobs=-1,
                    verbose=0,
                    random_state=42
                )
                random_search.fit(X_train, y_train)
                best_inner_model = random_search.best_estimator_
            else:
                pipeline.fit(X_train, y_train)
                best_inner_model = pipeline

            # Make predictions on the outer test fold
            y_pred = best_inner_model.predict(X_test)

            # Store predictions and true values
            all_preds = np.append(all_preds, y_pred)
            all_tests = np.append(all_tests, y_test)

            # Calculate evaluation metrics for the current fold
            accuracies.append(accuracy_score(y_test, y_pred))
            precisions.append(precision_score(y_test, y_pred, zero_division=0))
            recalls.append(recall_score(y_test, y_pred, zero_division=0))
            f1_scores.append(f1_score(y_test, y_pred, zero_division=0))

        # Store the average metrics across all outer folds
        avg_accuracy = np.mean(accuracies)
        avg_precision = np.mean(precisions)
        avg_recall = np.mean(recalls)
        avg_f1 = np.mean(f1_scores)

        cm = confusion_matrix(all_tests, all_preds)

        results_c[name] = {
            'accuracy': avg_accuracy,
            'precision': avg_precision,
            'recall': avg_recall,
            'f1': avg_f1,
            'confusion_matrix': cm
        }

        print(f"Average Accuracy: {avg_accuracy:.4f}")
        print(f"Average Precision: {avg_precision:.4f}")
        print(f"Average Recall: {avg_recall:.4f}")
        print(f"Average F1 Score: {avg_f1:.4f}")

        report = classification_report(all_tests, all_preds)
        print("Classification Report (Combined):\n", report)

        plt.figure(figsize=(6, 4))
        labels = ['No', 'Yes'] if task_name == 'Dropout' else ['Fail', 'Pass']
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                   xticklabels=labels, yticklabels=labels)
        plt.title(f'{name} {task_name} Confusion Matrix (Combined)')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()

        print("\n" + "="*50)

    return results_c

# Run both classification tasks
results_pass = run_classification_task(X_encoded, y_pass, 'Pass/Fail')
results_dropout = run_classification_task(X_encoded, y_dropout, 'Dropout')

# Print summary of results
def print_summary(results, task_name):
    print(f"\n{task_name} Results Summary:")
    print("="*50)
    print(f"{'Model':<25} {'Accuracy':<10} {'Precision':<10} {'Recall':<10} {'F1 Score':<10}")
    for name, metrics in results.items():
        print(f"{name:<25} {metrics['accuracy']:.4f}    {metrics['precision']:.4f}    {metrics['recall']:.4f}    {metrics['f1']:.4f}")

print_summary(results_pass, 'Pass/Fail')
print_summary(results_dropout, 'Dropout')

# Feature Importance for the best model (Random Forest) for both tasks
print("\n" + "="*50)
print("  FEATURE IMPORTANCE ANALYSIS  ")
print("="*50)

def plot_feature_importance(X, y, task_name):
    # Create a pipeline without PCA for interpretable feature importance
    pipeline_no_pca = ImbPipeline([
        ('scaler', StandardScaler()),
        ('smote', SMOTE(random_state=42)),
        ('clf', RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42))
    ])

    # Fit the pipeline on the full data
    pipeline_no_pca.fit(X, y)
    clf_no_pca = pipeline_no_pca.named_steps['clf']

    # Get feature importances
    feature_importances = clf_no_pca.feature_importances_
    feature_names = X.columns

    # Create a DataFrame for visualization
    importance_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': feature_importances
    }).sort_values(by='Importance', ascending=False)

    # Display the top 10 most important features
    print(f"\nTop 10 Features for {task_name}:")
    print(importance_df.head(10))

    # Visualization of Feature Importance
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Importance', y='Feature', data=importance_df.head(10))
    plt.title(f'Top 10 Important Features for {task_name}')
    plt.tight_layout()
    plt.show()

plot_feature_importance(X_encoded, y_pass, 'Pass/Fail')
plot_feature_importance(X_encoded, y_dropout, 'Dropout')
```

```python

# STEP 1: Create Summary DataFrame
# -----------------------------

def create_comparison(results, task_name):
    # Convert results to DataFrame
    metrics_df = pd.DataFrame(results).T.reset_index()
    metrics_df = metrics_df.rename(columns={
        'index': 'Model',
        'accuracy': 'Accuracy',
        'precision': 'Precision',
        'recall': 'Recall',
        'f1': 'F1 Score'
    })

    # Add task name column
    metrics_df['Task'] = task_name

    return metrics_df

# Create DataFrames for both tasks
pass_df = create_comparison(results_pass, 'Pass/Fail')
dropout_df = create_comparison(results_dropout, 'Dropout')

# Combine both DataFrames
combined_df = pd.concat([pass_df, dropout_df])

# Display summary
print("\n" + "="*60)
print("            MODEL PERFORMANCE SUMMARY")
print("="*60)
display(combined_df[['Task', 'Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score']])

# -----------------------------
# STEP 2: Visualize Comparison of Models
# -----------------------------
sns.set(style="whitegrid")
plt.figure(figsize=(16, 8))

# Melt the DataFrame for easier plotting
metrics_melted = pd.melt(
    combined_df,
    id_vars=['Model', 'Task'],
    value_vars=['Accuracy', 'Precision', 'Recall', 'F1 Score'],
    var_name='Metric',
    value_name='Score'
)

# Create subplots for each task
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Plot for Pass/Fail
sns.barplot(
    data=metrics_melted[metrics_melted['Task'] == 'Pass/Fail'],
    x='Model',
    y='Score',
    hue='Metric',
    palette='Set2',
    ax=axes[0]
)
axes[0].set_title('Pass/Fail Prediction Performance', fontsize=14)
axes[0].set_ylim(0, 1)
axes[0].set_xlabel('Model')
axes[0].set_ylabel('Score')
axes[0].tick_params(axis='x', rotation=30)

# Plot for Dropout
sns.barplot(
    data=metrics_melted[metrics_melted['Task'] == 'Dropout'],
    x='Model',
    y='Score',
    hue='Metric',
    palette='Set2',
    ax=axes[1]
)
axes[1].set_title('Dropout Prediction Performance', fontsize=14)
axes[1].set_ylim(0, 1)
axes[1].set_xlabel('Model')
axes[1].set_ylabel('')
axes[1].tick_params(axis='x', rotation=30)

# Adjust legend
axes[0].legend(title='Metrics', bbox_to_anchor=(1.05, 1), loc='upper left')
axes[1].get_legend().remove()

plt.tight_layout()
plt.show()
```

```python
# Add this at the end of your Colab notebook

# Save the preprocessed data and models
import joblib
import pickle

# Save the encoded features (we'll need this for the web app)
X_encoded.to_csv('student_features.csv', index=False)

# Save the target variables
y_pass.to_csv('y_pass.csv', index=False)
y_dropout.to_csv('y_dropout.csv', index=False)

# Save the best models for each task
# Let's train and save Random Forest models as they typically perform well
def save_best_model(X, y, task_name):
    pipeline = ImbPipeline([
        ('scaler', StandardScaler()),
        ('pca', PCA(n_components=0.95)),
        ('smote', SMOTE(random_state=42)),
        ('clf', RandomForestClassifier(n_estimators=300, max_depth=20, random_state=42))
    ])

    pipeline.fit(X, y)
    filename = f'best_model_{task_name.lower().replace("/", "_")}.pkl'
    joblib.dump(pipeline, filename)
    return filename

pass_model_file = save_best_model(X_encoded, y_pass, 'Pass/Fail')
dropout_model_file = save_best_model(X_encoded, y_dropout, 'Dropout')

# Save the column names for the web app
with open('feature_columns.pkl', 'wb') as f:
    pickle.dump(X_encoded.columns.tolist(), f)

# Download the files to your local machine
from google.colab import files
files.download('student_features.csv')
files.download('y_pass.csv')
files.download('y_dropout.csv')
files.download(pass_model_file)
files.download(dropout_model_file)
files.download('feature_columns.pkl')

print("All necessary files have been saved and downloaded.")
```
